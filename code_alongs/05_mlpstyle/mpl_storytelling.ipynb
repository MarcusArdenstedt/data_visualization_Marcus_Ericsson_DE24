# mplstyle and DRY for storytelling in matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
import duckdb


# in python script:
# Path(__file__).parent / "style/base.mplstyle"
plt.style.use("styles/base.mplstyle")

df = pd.read_csv("data/jobs_in_data_2024.csv")

df.head()
```

```python
df_job_categories = df["job_category"].value_counts()
df_job_categories
```

## visualization 


```python
from charts_utils import horizontal_bar_option, save_fig_from_ax
from constants import BLUE_1, GRAY_1

colors = [GRAY_1] * len(df_job_categories)
colors[:4] = [BLUE_1] * 4

ax = df_job_categories.plot(kind= "barh", color= colors)

ax= horizontal_bar_option(ax, title= "The top job categories within the data field 2024 are "
    + r"$\bf{data\ science},$"
    + "\n"
    + r"$\bf{data\ engineering},$"
    + r"$\bf{\ data\ analysis\ \text{and}\ machine\ learning}$",
                          xlabel= "# NUMBER OF JOBS", 
                          ylabel= "JOB CATEGORY")



save_fig_from_ax(ax, "figures/number_data_jobs.png")
```

```python
df_salary = (
    duckdb.query(
        """--sql
    SELECT 
        job_category, 
        CAST(AVG(salary_in_usd) AS INT) AS avg_salary_usd
    FROM 
        df
    GROUP BY 
        job_category
    ORDER BY avg_salary_usd DESC             
"""
    )
    .df()
    .set_index("job_category")
)

df_salary
```

```python
from charts_utils import thousands_formatter

colors = [GRAY_1] * len(df_salary)
colors[3] = BLUE_1


# important to use a pd.Series for color to work
ax = df_salary["avg_salary_usd"].plot(kind="barh", color= colors, width=0.6)

ax = horizontal_bar_option(ax, 
                        xlabel= "YEARLY SALARY IN USD", 
                        ylabel= "JOB CATEGORY",
                        title= r"$\bf{Data\ engineers}$"
    + " is both an extremely interesting field and also \none that tops the salary",)

ax = thousands_formatter(ax, axis= "x")

save_fig_from_ax(ax, "figures/data_jobs_salary.png")

```
