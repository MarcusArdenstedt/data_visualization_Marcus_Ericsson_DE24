```python
df_quantity_year = duckdb.query(
    """--sql
    SELECT
        Year,
        SUM(Quantity) AS Quantity
    FROM
        df
    GROUP BY
        Year
    ORDER BY
        Year
    ASC;
    """
    ).df()
df_quantity_year
```

```python
ax = df_quantity_year.iloc[:-1].plot(kind= "bar", x= "Year", y= "Quantity", figsize=(12,6))
ax.set_ylabel("Quantity")
ax.set_xticklabels(df_quantity_year["Year"].iloc[:-1], rotation=0)
ax.legend(["Tusen"])
ax.set(title="Antal bilar solda per Ã¥r och 2017 \nÃ¤r bort tagen fÃ¶r det var fÃ¶r lite data")

fig = ax.get_figure()
fig.tight_layout()
fig.savefig("figures/sold_cars_per_year")
```

## c)

```python
df_emission = df[["Year", "Month", "Avg_CO2"]]
df_emission = df_emission[["Year", "Month", "Avg_CO2"]].groupby("Year", as_index=False)[["Avg_CO2"]].sum().iloc[:-1]
df_emission
```

```python
import matplotlib.pylab as plt
fig, ax = plt.subplots(figsize= (12,6))
ax.bar(x= df_emission["Year"],
       height= df_emission["Avg_CO2"],
       width= 0.4)
ax.set_xticks(df_emission["Year"])
ax.set_xlabel("Year")
ax.set_ylabel("Total CO2 (g/km)")
ax.set_title("Total average CO2 per year")
plt.tight_layout()
plt.show()
```

```python
df_emission = df_emission.set_index("Year")
```

```python
print(list(df_emission.index))
```

```python
fig, ax = plt.subplots(figsize= (12,6))
ax.plot(df_emission.index, df_emission["Avg_CO2"], marker= 'o', markerfacecolor= 'red', markeredgecolor= 'black')

ax.set_xlim(df_emission.index.min())

ax.set_xticks(df_emission.index)
ax.set_xticklabels(df_emission.index.astype(str), rotation=0)

ax.set_ylabel("CO2 g/km")
ax.set_title("CO2 g/km per year (removed 2017)")
ax.set_xlabel("year")
plt.tight_layout()
plt.show()
```

## g)

```python
df_diesel_share = df[["Year", "Month", "Diesel_Share"]]
```

```python
df_diesel_share
```

```python
df_diesel_share.groupby("Year").agg({"Diesel_Share": "mean"}).reset_index()
```

```python
df_diesel_share = duckdb.query(
    """--sql
    SELECT
        Year,
        AVG(Diesel_Share) AS Average_diesel_Share
    FROM 
        df_diesel_share
    GROUP BY 
        Year
    ORDER BY 
        Year
    ASC;
    """
).df()
```

```python
df_diesel_share = df_diesel_share.iloc[:-1]
```

```python
df_diesel_share = df_diesel_share.set_index("Year")
```

```python
df_diesel_share
```

```python
fig, ax = plt.subplots(figsize= (12,5))
ax.plot(df_diesel_share.index, df_diesel_share["Average_diesel_Share"], linestyle= '--', color= 'blue')
ax.plot(df_diesel_share.index[1:], df_diesel_share["Average_diesel_Share"][1:], linestyle= '--', linewidth= 0.4, color= 'blue', marker = 'o', markersize= 4.5, markerfacecolor= 'red', markeredgecolor= 'black')

ax.set_xlim(df_diesel_share.index.min())
ax.set_xticks(df_diesel_share.index)
ax.set_xticklabels(df_diesel_share.index.astype(str), rotation= 0)
ax.set_yticks(np.arange(25, 90, 5))

ax.set_ylabel("diesel_share")
ax.set_xlabel("Year")
ax.set_title("Average diesel share per year from 2007-2016 (removed 2017)")




fig.savefig("figures/diesel_share_avg_per_year")
plt.tight_layout()
plt.show()
```
