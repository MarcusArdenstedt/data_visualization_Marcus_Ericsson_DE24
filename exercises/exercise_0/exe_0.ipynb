```python
df[["Year", "Quantity"]]
```

```python
import duckdb
```

```python
df_quantity_year = duckdb.query(
    """
    SELECT
        Year,
        SUM(Quantity) AS Quantity
    FROM
        df
    GROUP BY
        Year
    ORDER BY
        Year
    DESC;
    """
    ).df()
df_quantity_year
```

```python
df_quantity_year.plot(kind= "bar", x= "Year", y= "Quantity")
```
