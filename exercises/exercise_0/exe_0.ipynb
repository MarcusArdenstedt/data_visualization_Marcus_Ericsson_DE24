## B)

```python
df[["Year", "Quantity"]]
```

```python
import duckdb
```

```python
df_quantity_year = duckdb.query(
    """--sql
    SELECT
        Year,
        SUM(Quantity) AS Quantity
    FROM
        df
    GROUP BY
        Year
    ORDER BY
        Year
    ASC;
    """
    ).df()
df_quantity_year
```

```python
ax = df_quantity_year.iloc[:-1].plot(kind= "bar", x= "Year", y= "Quantity", figsize=(12,6))
ax.set_ylabel("Quantity")
ax.set_xticklabels(df_quantity_year["Year"].iloc[:-1], rotation=0)
ax.legend(["Tusen"])
ax.set(title="Antal bilar solda per år och 2017 \när bort tagen för det var för lite data")

fig = ax.get_figure()
fig.tight_layout()
fig.savefig("figures/sold_cars_per_year")
```

```python
df_quantity = df[["Year", "Month", "Quantity"]]
df_quantity
```

```python
df_quantity = duckdb.query(
    """--sql
    SELECT
        Year,
        SUM(Quantity) AS Quantity
    FROM 
        df_quantity 
    GROUP BY
        Year
    ORDER BY
        Year
    ASC;
    """
).df()
df_quantity
```

```python
df_quantity = df_quantity.iloc[:-1]
```

## c)

```python
df_emission = df[["Year", "Month", "Avg_CO2"]]
df_emission = df_emission[["Year", "Month", "Avg_CO2"]].groupby("Year", as_index=False)[["Avg_CO2"]].sum().iloc[:-1]
df_emission
```

```python
import matplotlib.pylab as plt
fig, ax = plt.subplots(figsize= (12,6))
ax.bar(x= df_emission["Year"],
       height= df_emission["Avg_CO2"],
       width= 0.4)
ax.set_xticks(df_emission["Year"])
ax.set_xlabel("Year")
ax.set_ylabel("Total CO2 (g/km)")
ax.set_title("Total average CO2 per year")
plt.tight_layout()
plt.show()
```
