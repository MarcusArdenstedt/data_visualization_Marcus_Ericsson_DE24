```python
import pandas as pd

df = pd.read_csv("data/co2_annmean_mlo.csv", comment= "#")
```

```python
df.set_index("year", inplace=True)
```

```python

from constant_2 import GRAY_1, GRAY_2, GRAY_3, RED_1, RED_2, RED_3, BLUE_1

ax = df.drop("unc", axis=1).plot(kind="line", linewidth= 2.2, color= RED_3)

#clutter
ax.spines[["top", "right"]].set_visible(False)
ax.legend().remove()
ax.tick_params(axis= "both", length= 0)

# contrast
ax.spines[["left", "bottom"]].set_color(GRAY_1) 
ax.tick_params("both", labelcolor= GRAY_2)

# description title and label - call to action (CTA)
ax.set_title("The annual mean of CO$_2$ emissions measured in Mauna Loa has \nincreased every year since 1959", 
             loc= "left", 
             color= GRAY_3, 
             pad= 10
             )
ax.set_ylabel("CO$_2$ MOLE FRACTON IN PPM", color= GRAY_2, loc= "top", rotation=90)
ax.set_xlabel("YEARS FROM 1959", color= GRAY_2, loc= "left")

fig = ax.get_figure()
fig.tight_layout()
fig.savefig("figures/co2_data_storytelling.png", dpi= 400)
```
